```python
data.fillna(data['Age'].median(),inplace=True)
```

```python
data.drop(['Name','Sex','Ticket','Embarked','Cabin'],axis=1,inplace= True)
```

```python
X=data.drop(['Survived'],axis=1)
print(X)
Y= data['Survived']
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.10, random_state=12)
```

```python
from sklearn.naive_bayes import GaussianNB
naive_chandni= GaussianNB()
naive_chandni.fit(X_train,Y_train)
```

```python
naive_predict=naive_chandni.predict(X_test)
print(naive_predict)
```

```python
from sklearn.metrics import accuracy_score
chandni_accuracy=accuracy_score(Y_test,naive_predict)
print(chandni_accuracy)
```

```python
import sklearn.metrics as metrics
chandni_cm=metrics.confusion_matrix(Y_test,naive_predict)
print(chandni_cm)
```
